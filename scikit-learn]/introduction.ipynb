```python
from sklearn.datasets import load_diabetes
```

```python
X, y = load_diabetes(return_X_y=True)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model = KNeighborsRegressor()
model.fit(X,y)
```

```python
pred = model.predict(X)
pred
```

```python
from sklearn import linear_model
model2 = linear_model.LinearRegression().fit(X,y)
```

```python
model2.predict(X)
```

```python
plt.scatter(pred,y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
pipe  = Pipeline([
    ('scale',StandardScaler()),
    ('model3',KNeighborsRegressor(n_neighbors=1))
])
```

```python
pipe.fit(X,y)
```

```python
pipe.predict(X)
```

```python
plt.scatter(pipe.predict(X),y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
pipe.get_params()
```

```python
model4 = GridSearchCV(estimator=pipe,
             param_grid={'model3__n_neighbors': range(1,10)},
             cv=3)
```

```python
model4.fit(X,y)
```

```python
pd.DataFrame(model4.cv_results_)
```
